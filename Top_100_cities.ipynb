```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as sp
```

```python
df=pd.read_csv('/content/Top 100 Worlds Largest Cities.csv')
df.head()
```

```python
df.describe()
df.info()
```

```python
df['Population (Est.)'] = df['Population (Est.)'].str.replace(',', '').astype(int)
df['Area (sq km)'] = df['Area (sq km)'].str.replace(',', '').astype(int)

population_density = df['Population (Est.)'] / df['Area (sq km)']
print(population_density)
```

```python
df_population_density=df[['City','Country']].copy()
df_population_density['Population_density']=population_density
df_population_density_sorted=df_population_density.sort_values(by='Population_density',ascending=False)
print(df_population_density_sorted)
```
