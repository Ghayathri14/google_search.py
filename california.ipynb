```python
from geopy.geocoders import Nominatim
geolocator=Nominatim(user_agent='geo_app')
# Get the first latitude and longitude from the scaled DataFrame
first_scaled_latitude = df['latitude'].iloc[0]
first_scaled_longitude = df['longitude'].iloc[0]

# Inverse transform the scaled coordinates to their original values
# The scaler expects a 2D array, so we reshape the single coordinate pair
original_coords = scaler.inverse_transform([[first_scaled_longitude, first_scaled_latitude]])
original_longitude = original_coords[0][0]
original_latitude = original_coords[0][1]

# Pass the original coordinates as a single string to the reverse method
location=geolocator.reverse(f"{original_latitude},{original_longitude}")

# Check if location was found before trying to access its address
if location:
    print(location.address)
else:
    print("Address not found for the given coordinates.")
```

```python
address=location.raw['address']
city=address.get('City','')
state=address.get('State','')
country=address.get('country','')
print(city,state,country)
```
